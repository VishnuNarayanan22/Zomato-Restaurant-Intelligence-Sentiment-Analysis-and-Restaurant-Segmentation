### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H₀: All cuisines have similar average ratings.

H₁: At least one cuisine has different ratings.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

cuisine_groups = []

for cuisine in restaurants['Cuisines'].dropna().unique():
    cuisine_restaurants = restaurants[restaurants['Cuisines'] == cuisine]['Name']

    cuisine_reviews = reviews[reviews['Restaurant'].isin(cuisine_restaurants)]

    ratings = cuisine_reviews['Rating'].dropna()

    if len(ratings) > 1:
        cuisine_groups.append(ratings)

# Perform ANOVA
f_statistic, p_value = f_oneway(*cuisine_groups)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

I used the One-Way ANOVA (Analysis of Variance) test to obtain the p-value

##### Why did you choose the specific statistical test?

I chose the One-Way ANOVA test because the variable Cuisines is categorical (different types of cuisines), while Rating is a numerical variable. ANOVA is suitable when we want to compare the means of a numerical variable across multiple groups to determine if the differences between them are statistically significant.
